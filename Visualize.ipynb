```python
import scipy.io
import matplotlib.pyplot as plt
```

```python
test_data = scipy.io.loadmat("synthetic_data_test_2_gaussians.mat")["test_data"]
test_label = scipy.io.loadmat("synthetic_data_test_2_gaussians.mat")["test_label"]
```

```python
import numpy as np

# Prepare the data for each class
num_samples_per_class = 10
class_1_data = test_data[:num_samples_per_class]
class_2_data = test_data[1000:1000 + num_samples_per_class]

# Plot the first 10 samples of class 1
fig, ax = plt.subplots(figsize=(10, 6))
x_values = np.arange(100)
for i in range(num_samples_per_class):
    ax.plot(x_values, class_1_data[i], label=f'Sample {i+1}')
ax.set_xlabel('Sample Times')
ax.set_ylabel('Value')
plt.title('First 10 Class 1 Train Data Samples Visualization')
ax.legend()
plt.show()

# Plot the first 10 samples of class 2
fig, ax = plt.subplots(figsize=(10, 6))
for i in range(num_samples_per_class):
    ax.plot(x_values, class_2_data[i], label=f'Sample {i+1}')
ax.set_xlabel('Sample Times')
ax.set_ylabel('Value')
plt.title('First 10 Class 2 Train Data Samples Visualization')
ax.legend()
plt.show()
```

Visualize the warping functions *gamma*

```python
# Load gamma values
gamma_data = scipy.io.loadmat("./generated_gamma_github.mat")["generated_gamma"]

# Prepare the gamma values for each class
num_samples_per_class = 10
gamma_class_1 = gamma_data[:num_samples_per_class]
gamma_class_2 = gamma_data[1000:1000 + num_samples_per_class]

# Plot the first 10 gamma values of class 1
fig, ax = plt.subplots(figsize=(10, 6))
x_values = np.arange(100)
for i in range(num_samples_per_class):
    ax.plot(x_values, gamma_class_1[i], label=f'Sample {i+1}')
ax.set_xlabel('Input Time Index')
ax.set_ylabel('Output Time Index')
plt.title('First 10 Warping Functions (Gamma) for Class 1 Test Data Samples')
ax.legend()
plt.show()

# Plot the first 10 gamma values of class 2
fig, ax = plt.subplots(figsize=(10, 6))
for i in range(num_samples_per_class):
    ax.plot(x_values, gamma_class_2[i], label=f'Sample {i+1}')
ax.set_xlabel('Input Time Index')
ax.set_ylabel('Output Time Index')
plt.title('First 10 Warping Functions (Gamma) for Class 2 Test Data Samples')
ax.legend()
plt.show()
```
